## Introduction To Activation Functions

Neural Networks

Logistic regression models learn a set of weights that impact the linear combination phase and then are fed through a single nonlinear function (the sigmoid function). In this mission, we'll dive into the most commonly used activation functions. The three most commonly used activation functions in neural networks are:

* the sigmoid function
* the ReLU function
* the tanh function

## ReLU Activation Function

We'll start by introducing the **ReLU** activation function, which is a commonly used activation function in neural networks for solving regression problems. ReLU stands for **rectified linear unit** and is defined as follows:

$\operatorname{ReLU}(x) = \max(0,x)$

The $\max(0,x)$ function call returns the maximum value between $0$ and $x$.

This means that:

* when $x$ is less than $0$, the value $0$ is returned.
* when $x$ is greater than $0$, the value $x$ is returned

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
def relu(x):
    return np.maximum(x, 0)
```

```python
x = np.linspace(-2, 2, 20)
```

```python
relu_y = relu(x)
print(x)
print(relu_y)
```

```python
relu([-1,3,0,2,-2])
```

```python
relu(-3)
```

```python
plt.plot(x, relu_y);
```

## Trigonometric Functions

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)

tan_y = np.tan(x)

print(x)
print(tan_y)

plt.plot(x, tan_y);
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)

y = np.sin(x)

plt.plot(x,y);
```

## Hyperbolic Tangent Function

```python
x = np.linspace(-40, 40, 100)

tanh_y = np.tanh(x)

print(x)
print(tanh_y)

plt.plot(x, tanh_y);
```

## Summary

* Both the **sigmoid** and the **tanh** functions are commonly used in neural networks for **classification** tasks.

* The **ReLU** function, on the other hand, is known to be more effective in **regression** tasks.
