## Introduction To Derivatives

Let's first start by rewriting the slope equation to the more common form. Instead of using $x_1$ and $x_2$ in the slope equation, let's use $x$ and $x+h$. This version of the slope equation is how most textbooks and resources refer to slope:

$\lim_{h \to 0} \frac{f(x + h) - f(x)}{h}$

![Jupyter](./slope_h_form.gif)

## Differentiation

$f(x) = -(x)^2 + 3x - 1$

$\lim_{h \to 0} \frac{(-(x + h)^2 + 3(x + h) - 1) - (-(x)^2 + 3x - 1)}{h}$

$\lim_{h \to 0} -2x - h + 3$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 6, 110)
y1 = -2*x + 3
y2 = -1*(x**2) + 3*x -1



plt.plot(x,y1)
plt.plot(x,y2)
plt.vlines(1.5,-40, 10)
plt.show()
```

## Extreme Values

* When the slope transitions from `negative to positive` at a point, it can be a `minimum` value. 
* When the slope transitions from `positive to negative` at a point, it can be a `maximum` value.

## Power Rule

$f(x) = x^{r}$

$f'(x) = r x^{r-1}$

## Linearity Of Differentiation

Linearity of differentiation consists of 2 rules. 

First is the sum rule:

$\frac {d}{dx} [f(x) + g(x)] = \frac {d}{dx} [f(x)] + \frac {d}{dx} [g(x)]$

Second is the constant factor rule, which lets us pull out constants from the derivative:

$\frac {d}{dx} [cf(x)] = c\frac {d}{dx} [f(x)]$
