## Math Behind Slope

```python
def slope(x1,x2,y1,y2):
    ## Finish the function here
    slope = (y2-y1)/(x2-x1)
    return slope
slope_one = slope(0,4,1,13)
slope_two = slope(5,-1,16,-2)
```

## Secant Lines

```python
import seaborn
seaborn.set(style='darkgrid')

def draw_secant(x_values):
    x = np.linspace(-20,30,100)
    y = -1*(x**2) + x*3 - 1
    plt.plot(x,y)
    
    # Add your code here.
    x1 = x_values[0]
    y1 = -1*(x1**2) + x1*3 - 1
    x2 = x_values[1]
    y2 = -1*(x2**2) + x2*3 - 1
    slope = (y2-y1)/(x2-x1)
    intercept = y1 - slope*x1
    
    y_secant = x*slope+intercept
    plt.plot(x, y_secant, c='green')
    
    plt.show()
```

```python
draw_secant([3,5])
draw_secant([3,10])
draw_secant([3,15])
```
