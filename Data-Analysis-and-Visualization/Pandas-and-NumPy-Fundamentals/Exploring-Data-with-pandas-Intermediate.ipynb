## Using Loops with pandas

```python
top_employer_by_country = {}

countries = f500['country'].unique()
```

```python
for country in countries:
    country_records = f500[f500['country'] == country]
    top_employer = country_records.sort_values('employees', ascending=False).head(1).iloc[0]['company']
    top_employer_by_country[country] = top_employer
```

```python
top_employer_by_country
```
