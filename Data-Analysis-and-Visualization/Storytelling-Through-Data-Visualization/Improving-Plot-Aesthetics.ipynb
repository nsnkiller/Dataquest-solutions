Randal compiled this data set to explore the gender gap in STEM fields, which stands for science, technology, engineering, and mathematics.
 we'll explore how we can communicate the nuanced narrative of gender gap using effective data visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
women_degrees.info()
```

```python
plt.plot(women_degrees['Year'], women_degrees['Biology']);
```

## Visualizing The Gender Gap

```python
plt.plot(women_degrees['Year'], women_degrees['Biology'], color='blue', label='Women')
plt.plot(women_degrees['Year'], 100 - women_degrees['Biology'], color='green', label='Men')
plt.title('Percentage of Biology Degrees Awarded By Gender')
plt.legend(loc='upper right');
```

## Data-Ink Ratio

To improve the data-ink ratio, let's make the following changes to the plot we created in the last step:

* Remove all of the axis tick marks.
* Hide the spines, which are the lines that connects the tick marks, on each axis.

![Jupyter](./data-ink.gif)

## Hiding Tick Marks

By default, matplotlib displays the tick marks on all four sides of the plot. Here are the four sides for a standard line chart:

* The left side is the y-axis.
* The bottom side is the x-axis.
* The top side is across from the x-axis.
* The right side is across from the y-axis.

```python
fig,ax = plt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], color='blue', label='Women')
ax.plot(women_degrees['Year'], 100 - women_degrees['Biology'], color='green', label='Men')
ax.tick_params(bottom="off", top="off", left="off", right="off")
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc='upper right')
plt.show()
```

## Hiding Spines

```python
fig,ax = plt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], color='blue', label='Women')
ax.plot(women_degrees['Year'], 100 - women_degrees['Biology'], color='green', label='Men')
ax.tick_params(bottom="off", top="off", left="off", right="off")


ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)

# for key,spine in ax.spines.items():
#     spine.set_visible(False)

ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc='upper right')
plt.show()
```

## Comparing Gender Gap Across Degree Categories

```python
category_feature = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
n_rows = 2
n_cols = 2
category_feature = np.array(category_feature).reshape(n_rows,n_cols)

fig,axes = plt.subplots(n_rows, n_cols, figsize=(12,12))
for row in range(n_rows):
    for col in range(n_cols):
        category = category_feature[row][col]
        ax = axes[row][col]
        
        ax.plot(women_degrees['Year'], women_degrees[category], color='blue', label='Women')
        ax.plot(women_degrees['Year'], 100-women_degrees[category], color='green', label='Men')
        ax.set_xlim(1968,2011)
        ax.set_ylim(0,100)
        ax.tick_params(bottom=False, top=False, left=False, right=False)
        for key,spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_title(category)
        if (row == n_rows-1) and (col == n_cols - 1):
            ax.legend(loc='upper right')
plt.show()
```
