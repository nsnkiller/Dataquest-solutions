## Pandas, Histograms

```python
features_explore = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
bins = range(10,25,5)

for feature in features_explore:
    for bin in bins:
        plt.figure()
        ax = recent_grads[feature].plot(kind='hist', bins=bin)
        ax.set_title(feature + ' Distribution with bins:'+ str(bin));
```

## Pandas, Bar Plots

```python
ax = recent_grads[:10].plot.bar(x='Major', y='ShareWomen', figsize=(10,8))
ax.set_ylabel('Women Share')
ax = recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', figsize=(10,8))
ax.set_ylabel('Women Share');
```
