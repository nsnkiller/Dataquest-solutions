## Introduction To The Data

```python
import pandas as pd
```

```python
unrate = pd.read_csv('unrate.csv')
unrate.info()
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
unrate.head(12)
```

## Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
# %matplotlib inline

plt.plot()
plt.show()
```

## Adding Data

```python
plt.figure(figsize=(16,8))
plt.plot(unrate[unrate.DATE.dt.year==1948]['DATE'], unrate[unrate.DATE.dt.year==1948]['VALUE'])
plt.xticks(unrate[unrate.DATE.dt.year==1948]['DATE'])
plt.show()
```

## Fixing Axis Ticks

```python
plt.plot(unrate[unrate.DATE.dt.year==1948]['DATE'], unrate[unrate.DATE.dt.year==1948]['VALUE'])
plt.xticks(unrate[unrate.DATE.dt.year==1948]['DATE'])
plt.xticks(rotation=90)
plt.show()
```

## Adding Axis Labels And A Title

```python
plt.plot(unrate[unrate.DATE.dt.year==1948]['DATE'], unrate[unrate.DATE.dt.year==1948]['VALUE'])
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```
