## Matplotlib Classes

what pyplot was automatically storing under the hood when we create a single plot:

* a container for all plots was created (returned as a Figure object)
* a container for the plot was positioned on a grid (the plot returned as an Axes object)
* visual symbols were added to the plot (using the Axes methods)

The Axes object acts as its own container for the various components of the plot, such as:

* values on the x-axis and y-axis
* ticks on the x-axis and y-axis
* all visual symbols, such as:

    * markers
    * lines
    * gridlines

This will create a grid, 2 rows by 1 column, of plots

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1) # row, column, postion
ax2 = fig.add_subplot(2,1,2)
plt.show()
```

## Adding Data

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

data_1948 = unrate[unrate['DATE'].dt.year==1948]
data_1949 = unrate[unrate['DATE'].dt.year==1949]

ax1.plot(data_1948['DATE'], data_1948['VALUE'])
ax2.plot(data_1949['DATE'], data_1949['VALUE'])

plt.show()
```

## Formatting And Spacing

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

data_1948 = unrate[unrate['DATE'].dt.year==1948]
data_1949 = unrate[unrate['DATE'].dt.year==1949]

ax1.plot(data_1948['DATE'], data_1948['VALUE'])
ax1.set_title('Monthly Unemployment Rate, 1948')
ax2.plot(data_1949['DATE'], data_1949['VALUE'])
ax2.set_title('Monthly Unemployment Rate, 1949')
plt.show()
```

## Comparing Across More Years

```python
fig = plt.figure(figsize=(12,12))

for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    data = unrate[unrate['DATE'].dt.year==(1948+i)]
    ax.plot(data['DATE'], data['VALUE'])
    #ax.set_title('Monthly Unemployment Rate, {}'.format(1948+i))

plt.show()
```

## Overlaying Line Charts

```python
unrate['MONTH'] = unrate['DATE'].dt.month
```

```python
fig = plt.figure(figsize=(6,3))
for i in range(2):
    data = unrate[unrate['DATE'].dt.year==(1948+i)]
    plt.plot(data['MONTH'], data['VALUE'])
plt.show()
```

## Adding More Lines

```python
fig = plt.figure(figsize=(10,6))

colors ={1948:'red', 1949:'blue', 1950:'green', 1951:'orange', 1952:'black'}
for i in range(5):
    data = unrate[unrate['DATE'].dt.year==(1948+i)]
    plt.plot(data['MONTH'], data['VALUE'], c=colors[1948+i])

plt.show()
```

## Adding A Legend

```python
fig = plt.figure(figsize=(10,6))

colors ={1948:'red', 1949:'blue', 1950:'green', 1951:'orange', 1952:'black'}
for i in range(5):
    data = unrate[unrate['DATE'].dt.year==(1948+i)]
    plt.plot(data['MONTH'], data['VALUE'], c=colors[1948+i], label=str(i+1948))

plt.legend(loc = 'upper left')
plt.show()
```

## Final Tweaks

```python
fig = plt.figure(figsize=(10,6))

colors ={1948:'red', 1949:'blue', 1950:'green', 1951:'orange', 1952:'black'}
for i in range(5):
    data = unrate[unrate['DATE'].dt.year==(1948+i)]
    plt.plot(data['MONTH'], data['VALUE'], c=colors[1948+i], label=str(i+1948))

plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.legend(loc = 'upper left')
plt.show()
```
